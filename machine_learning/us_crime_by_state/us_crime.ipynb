```python
from numpy import random   
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv('/Users/jesshuang/Documents/GitHub/jess_window/machine_learning/us_crime_by_state/US_violent_crime.csv')
data.head()
```

```python
data.describe()
```

```python
d = pd.DataFrame(data,columns=['Rape'])
d.describe()
```

```python
from pandas.plotting import scatter_matrix

_=scatter_matrix(data, figsize=(5,5))
```

```python
Q1=data['Rape'].quantile(0.25)
Q3=data['Rape'].quantile(0.75)
IQR=Q3-Q1
Upper=Q3+1.5*IQR
Lower=Q1-1.5*IQR
print('Q3=',Q3,'Q1=',Q1,'IQR=',IQR,'Upper=',Upper,'Lower=',Lower)
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(d,showmeans=True)
plt.title('Rape Rate')
plt.show()
```

```python
sns.distplot(d)
plt.show()
```

```python
sns.displot(d)
plt.show()
```

```python
sns.histplot(d)
plt.show()
```

```python
#skewness and kurtosis
print("偏態 Skewness: {:.2f}".format(d['Rape'].skew()))
print("峰度 Kurtosis: {:.2f}".format(d['Rape'].kurt()))
```

```python
exp_data = np.power(d['Rape'], 2)
exp_data = exp_data.replace([np.inf, -np.inf, -0], 0)
sns.distplot(exp_data)
plt.show()
```

```python
#使用10*10均勻分配資料
np.random.seed(0)
rdata=np.random.rand(10,10)
sns.heatmap(rdata)
```

```python
#設定顏色範圍 (vmin, vmax:設定顏色範圍)
sns.heatmap(rdata, vmin=0.5, vmax=0.8)
```

```python
sns.heatmap(rdata, center=0.5)
```

```python
from numpy.random.mtrand import normal
#抽取標準常態分配10*10的數據,以0為中心點

np.random.seed(0)

normal_data = np.random.randn(10,10) #random.randn(n,n) 是以常態分配為依據

sns.heatmap(normal_data, center=0)
```
