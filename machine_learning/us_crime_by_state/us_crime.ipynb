```python
from numpy import random   
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv('/Users/jesshuang/Documents/GitHub/jess_window/machine_learning/us_state_crime/US_violent_crime.csv')
data.head()
```

```python
data.describe()
```

```python
d = pd.DataFrame(data,columns=['Rape'])
d.describe()
```

```python
from pandas.plotting import scatter_matrix

_=scatter_matrix(data, figsize=(5,5))
```

```python
Q1=data['Rape'].quantile(0.25)
Q3=data['Rape'].quantile(0.75)
IQR=Q3-Q1
Upper=Q3+1.5*IQR
Lower=Q1-1.5*IQR
print('Q3=',Q3,'Q1=',Q1,'IQR=',IQR,'Upper=',Upper,'Lower=',Lower)
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(d,showmeans=True)
plt.title('Rape Rate')
plt.show()
```

```python
sns.distplot(d)
plt.show()
```

```python
sns.displot(d)
plt.show()
```

```python
sns.histplot(d)
plt.show()
```

```python
#skewness and kurtosis
print(" Skewness: {:.2f}".format(d['Rape'].skew()))
print(" Kurtosis: {:.2f}".format(d['Rape'].kurt()))
```

```python
exp_data = np.power(d['Rape'], 2)
exp_data = exp_data.replace([np.inf, -np.inf, -0], 0)
sns.distplot(exp_data)
plt.show()
```
